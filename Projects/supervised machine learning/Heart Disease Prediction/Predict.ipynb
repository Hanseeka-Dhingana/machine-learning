# Heart Disease Prediction using Logistic Regression Model
Heart disease prediction using multiple features with a `logistic regression model` for accurate risk assessment.

## Step 1: Import all required libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
```

## Step 2: Load the Dataset perform Basic EDA and Preprocessing

```python
df = pd.read_csv("Heart_Disease_Prediction.csv")
df
```

```python
df.describe().T              #return the mean, standard derivation, quantile, percentile, deciles of the numerical columns
```

```python
df.info()          # return all the information about dataset
```

```python
df.shape             # shape attribute tell the dimension of the dataset
```

```python
df.isna().sum()            # check the null values
```

## Step 3: Convert Categorical data to numerical 

```python
# create object of labelEncoder()
le = LabelEncoder()

# Convert categorical to numerical 
df["Heart Disease"] = le.fit_transform(df["Heart Disease"])         # Absence → 0, Presence → 1

# Check after converting the values.
df.head()
```

# Step 4: Split data into 80% training, 20% testing

```python
X = df.drop("Heart Disease", axis = 1)        # features
Y = df["Heart Disease"]                      # Target


x_train, x_test, y_train,  y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

## Step 5: Train the model 

```python
# create an object
lr = LogisticRegression(max_iter=1000)

#train the model
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)

lr.score(x_test, y_test)
r2_score(y_test, y_pred)
```
